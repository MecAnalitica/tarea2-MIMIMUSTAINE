Muchos de los fenómenos que podemos describir respecto al parámetro **tiempo** son ejemplos de movimientos periódicos: oscilaciones de un reloj de péndulo, vibraciones de cuerdas, etc. Podemos relacionar los fenómenos anteriores al concluir que su patrón de movimiento se repite una y otra vez. 

Sin embargo, muchos de los fenómenos físicos no son periódicos, en algunos, notamos cierta tendencia a serlo, aún cuando interacciones externas lo motivan a no serlo. Para esta clase de fenómenos utilizamos una metodología análoga para su análisis considerando las interacciones externas, un ejemplo de éstos es el **oscilador forzado amortiguado**.

El **oscilador forzado amortiguado** está sujeto a una fuerza periódica debida a un factor externo. Esta fuerza periódica obliga al oscilador a vibrar a su frecuencia, iniciando con su frecuencia en la primera oscilación. Una de estas frecuencias (oscilador o fuerza) cede mientras que la otra persiste, este movimiento se conoce como **transitorio**, eventualmente el sistema adquiere un **movimiento estable**. La *fuerza propulsora* se representa como

$$F = F_{o}e^{i \omega t}$$ (I)

entonces, el movimiento del oscilador se expresa de la manera siguiente:

$$m \ddot x + c \dot x + kx = F_{o}e^{i \omega t}$$ (II)

Dado que la variable $x$ es real, por la identidad de Euler podemos simplificar la expresión anterior comparando las partes real e imginaria para encontrar una relación entre las variables. Una solución propuesta a la ecuación diferencial anterior es $x(t) = Ae^{i(\omega t - \phi)}$ donde $A, \phi$ son parámetros a ser determinados. Sustituyendo en la ecuación diferencial obtenemos la siguiente expresión:

$$-m \omega^{2}A + i \omega cA + kA = F_{o}e^{i \phi} = F_{o}(cos(\phi) + isen(\phi))$$ (III)

Al igualar las partes real e imaginaria, se tiene que

$$tan(\phi) = \dfrac{c\omega}{k - m\omega^{2}}$$ (IV)

donde $\phi$ es el ángulo de fase. Con la identidad pitagórica $sen^{2}(\phi) + cos^{2}(\phi) = 1$ y la expresión (III), en términos de las abreviaciones $\omega_{o}^{2} = k/m$ y $\gamma = c/2m$, se sigue que

$$tan(\phi) = \dfrac{2\gamma \omega}{\omega_{o}² - \omega²}$$ (V)

$$A(\omega) = \dfrac{F_{o}/m}{\sqrt{(\omega_{o}^{2} - \omega^{2})^2 + 4\gamma² \omega²}}$$ (VI)

Un ejemplo de *oscilador amortiguado forzado* es el **péndulo forzado amortiguado** cuya expresión que refiere a su movimiento equivale a la serie mostrada anteriormente. Ésto implica que su expresión de movimiento tiene la siguiente forma:

$$\theta(t) = \theta_{HOM} + \dfrac{F_{o}/m}{\sqrt{(\omega_{o}^{2} - \omega^{2})^2 + 4\gamma² \omega²}}cos(\omega t -\phi)$$ (VII)

El término $\theta_{HOM} = \delta e^{-\gamma t}$, $\delta \in \mathbb{R}$, por lo que la expresión puede aproximarse únicamente al segundo término. Concluida la deducción de la expresión del movimiento para un *péndulo forzado amortiguado*, es necesario analizar gráficamente una serie de puntos, es decir, caracterizar.

Para determinar las condiciones iniciales del sistema, considérese la solución de la ecuación diferencial correspondiente al límite lineal de un oscilador amortiguado forzado, la cual tiene la siguiente forma:

$$\theta(t) = A^{(I)}e^{-\gamma t}cos(\omega^{(I)}t - \delta^{(I)}) + A^{(II)}cos(\omega^{(II)}t - \delta^{(II)})$$

donde en la parte derecha de la igualdad se representa la suma de las soluciones **homogénea $(\theta_{H})$** y **particular $(\theta_{P})$** respectivamente. Se analizarán las condiciones considerando $\omega^{(I)}, \omega^{(II)}$ conocidas. El análisis se realizará por separado.

**Ejemplo**

**Solución homogénea $(\theta_{H})$**

Con $\theta_{H}(0) = q_{1}$, $\dot \theta_{H}(0) = q_{2}$ se determinan a continuación el valor de los parámetros $A^{(I)}, \delta^{(I)}$, además consideremos que el oscilador parte del reposo, así $q_{2} = 0$.

$\theta_{H}(t) = A^{(I)}e^{-\gamma t}cos(\omega^{(I)}t - \delta^{(I)})$, $\theta_{H}(0) = A^{(I)}cos( - \delta^{(I)}) = A^{(I)}cos(\delta^{(I)}) = q_{1}$, $(i)$

$\dot \theta_{H}(t) = - A^{(I)} \gamma e^{-\gamma t}cos(\omega^{(I)}t - \delta^{(I)}) - A^{(I)} \omega^{(I)} e^{-\gamma t}sen(\omega^{(I)}t - \delta^{(I)})$, $\dot \theta_{H}(0) = - A^{(I)} \gamma cos(- \delta^{(I)}) - A^{(I)} \omega^{(I)} sen(- \delta^{(I)}) = - A^{(I)} \gamma cos(\delta^{(I)}) + A^{(I)} \omega^{(I)} sen(\delta^{(I)}) = q_{2} = 0$ $(ii)$

De la expresión $(ii)$, se determina que $\delta^{I} = arctan(\dfrac{\gamma}{\omega^{I}})$ y, por consecuente, que $A^{(I)} = \dfrac{q_{1}}{cos\delta^{(I)}}$, siendo $\gamma = 0.05$, $\omega^{(I)} = \sqrt{\omega_{o}² - \gamma²}$, por tanto, para las dos condiciones iniciales sugeridas, los valores de los parámetros serán los siguientes, con $q_{1} = \dfrac{\pi}{30}$, $q_{1} = \dfrac{\pi}{50}$:

$$\delta^{(I)}_{1,2} = 0.050021, A^{(I)}_{1} = 0.104720, A^{(I)}_{2} = 0.062832$$

**Solución particular $(\theta_{P})$**

Con $\theta_{P}(0) = p_{1}$, $\dot \theta_{H}(0) = p_{2}$ se determinan a continuación el valor de los parámetros $A^{(II)}, \delta^{(II)}$, además consideremos que el oscilador parte del reposo, así $p_{2} = 0$.

$\theta_{P}(t) = A^{(II)}cos(\omega^{(II)}t - \delta^{(II)})$, $\theta_{P}(0) = A^{(II)}cos( - \delta^{(II)}) = A^{(II)}cos(\delta^{(II)}) = p_{1}$, $(iii)$

$\dot \theta_{P}(t) = - A^{(II)} \omega^{(II)}sen(\omega^{(II)}t - \delta^{(II)})$, $\dot \theta_{P}(0) =- A^{(II)} \omega^{(II)} sen(- \delta^{(II)}) = A^{(II)} \omega^{(II)} sen(\delta^{(II)}) = q_{2} = 0$ $(iv)$

De la expresión $(iv)$, se determina que $\delta^{II} = n\pi$, en este caso $n = 0$, $A^{(II)} = \dfrac{q_{1}}{cos\delta^{(II)}}$, siendo $\gamma = 0.05$, $\omega^{(II)} = 2$, por tanto, para las dos condiciones iniciales sugeridas, los valores de los parámetros serán los siguientes, con $q_{1} = \dfrac{\pi}{60}$, $q_{1} = \dfrac{\pi}{25}$:

$$\delta^{(II)}_{1,2} = 0, A^{(II)}_{1} = \dfrac{\pi}{60}, A^{(II)}_{2} = \dfrac{\pi}{25}$$

Por tanto, las condiciones iniciales de la expresión del oscilador son las siguientes:

$$\theta(0) = \dfrac{\pi}{20}, \dot \theta(0) = 0$$
$$\theta(0) = \dfrac{3}{50}\pi, \dot \theta(0) = 0$$

Para evitar el desarrollo analítico, puede procederse a la implementación de la determinación de condiciones iniciales en el método numérico.

```python
import matplotlib. pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint

#Para condiciones iniciales.
def As(q,delta):
    return (q / (np.cos(delta)))

def del1(gamma,w1):
    return np.arctan(gamma / w1)              #Sólo para q2 = 0

def del2(q1,q2,wf):
    return np.arctan(q2 / (q1 * wf))

#Funciones.
def the(t,gamma,Auno,wuno,delta1,Ados,wfuerza,delta2):
    return (Auno * np.exp(- gamma * t) * np.cos ((wuno * t) - delta1)) + (Ados * np.cos((wfuerza * t) - delta2))

def ome(t,gamma,Auno,wuno,delta1,Ados,wfuerza,delta2):
    return (- (Auno * gamma * np.exp(-gamma * t) * np.cos(wuno * t - delta1) - (Auno * wuno * np.exp(-gamma * t) * np.sin(wuno * t - delta1)) - (Ados * wfuerza * np.sin(wfuerza * t - delta2))))

t = np.linspace(0,200,100000)
omo = 1.0
gamma = 0.05
w1 = np.sqrt(pow(omo,2) - pow(gamma,2))
wf = 1.5
A = 1.0
b = 0
A1 = []
A2 = []
q = [np.pi / 3, np.pi / 5, 0, 0]
p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular
delta1 = del1(gamma,w1)
delta2 = []

for i in range (2):
    delta2.append(del2(p[i],p[i+2],wf))

for i in range (2):
    A1.append(As(q[i],delta2[0]))
    A2.append(As(q[i],delta2[1]))

#Inciso C
theta1 = the(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
                  
plt.figure(figsize=(20,5))
plt.plot(t,theta1,'k',label='$\\theta_{1}(t)$', linewidth=2)
plt.plot(t,theta2,'darkred',label='$\\theta_{2}(t)$', linewidth=2)
plt.title('Ángulo en función del tiempo (dos condiciones iniciales), $A_{H} \\neq 0$')
plt.ylabel('Ángulo $(\\theta)$')
plt.xlabel('Tiempo $(t)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

plt.figure(figsize=(20,5))
plt.plot(t,omega1,'k',label='$\\dot \\theta)_{1}(t)$', linewidth=2)
plt.plot(t,omega2,'darkred',label='$\\dot \\theta)_{2}(t)$', linewidth=2)
plt.title('Frecuencia angular en función del tiempo (dos condiciones iniciales), $A_{H} \\neq 0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Tiempo $(t)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Si A1 = 0

theta1 = the(t,gamma,b,w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,b,w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,b,w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,b,w1,delta1,A,wf,delta2[1])

plt.figure(figsize=(20,5))
plt.plot(t,theta1,'k',label='$\\theta_{1}(t)$', linewidth=2)
plt.plot(t,theta2,'darkred',label='$\\theta_{2}(t)$', linewidth=2)
plt.title('Ángulo en función del tiempo (dos condiciones iniciales), $A_{H} = 0$')
plt.ylabel('Ángulo $(\\theta)$')
plt.xlabel('Tiempo $(t)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()
                  
plt.figure(figsize=(20,5))
plt.plot(t,omega1,'k',label='$\\dot \\theta_{1}(t)$', linewidth=2)
plt.plot(t,omega2,'darkred',label='$\\dot \\theta)_{2}(t)$', linewidth=2)
plt.title('Frecuencia angular en función del tiempo (dos condiciones iniciales), $A_{H} = 0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Tiempo $(t)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Inciso D
q = [5 * np.pi / 6, 4 * np.pi / 5, 0, 0]
p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular

for i in range (2):
    delta2.append(del2(p[i],p[i+2],wf))

for i in range (2):
    A1.append(As(q[i],delta2[0]))
    A2.append(As(q[i],delta2[1]))

theta1 = the(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} \\neq 0$, $\\gamma = 0.0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Si A1 = 0
theta1 = the(t,gamma,b,w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,b,w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,b,w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,b,w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} = 0$, $\\gamma = 0.0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Inciso E
gamma1 = 0.0
w1 = np.sqrt(pow(omo,2) - pow(gamma1,2))
q = [np.pi / 3, np.pi / 5, 0, 0]
p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular
delta1 = del1(gamma1,w1)
delta2 = []

for i in range (2):
    delta2.append(del2(p[i],p[i+2],wf))

for i in range (2):
    A1.append(As(q[i],delta2[0]))
    A2.append(As(q[i],delta2[1]))

theta1 = the(t,gamma,A1[0],w1,delta1,A2[0],wf,delta2[0])
theta2 = the(t,gamma,A1[1],w1,delta1,A2[1],wf,delta2[1])
omega1 = ome(t,gamma,A1[0],w1,delta1,A2[0],wf,delta2[0])
omega2 = ome(t,gamma,A1[1],w1,delta1,A2[1],wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} \\neq 0$, $A \\neq 1$, $\\gamma = 0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Si A1 = 0

theta1 = the(t,gamma,b,w1,delta1,A2[0],wf,delta2[0])
theta2 = the(t,gamma,b,w1,delta1,A2[1],wf,delta2[1])
omega1 = ome(t,gamma,b,w1,delta1,A2[0],wf,delta2[0])
omega2 = ome(t,gamma,b,w1,delta1,A2[1],wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} = 0$, $A \\neq 1$, $\\gamma = 0.0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Inciso F
#Ahora con diferentes gammas

#Con gamma = 0
gamma = 0.0
w1 = np.sqrt(pow(omo,2) - pow(gamma,2))
q = [np.pi / 3, np.pi / 5, 0, 0]
p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular
delta1 = del1(gamma1,w1)
delta2 = []

for i in range (2):
    delta2.append(del2(p[i],p[i+2],wf))

for i in range (2):
    A1.append(As(q[i],delta2[0]))
    A2.append(As(q[i],delta2[1]))

theta1 = the(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} \\neq 0$, $\\gamma = 0.0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Si A1 = 0

theta1 = the(t,gamma,b,w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,b,w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,b,w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,b,w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} = 0$, $\\gamma = 0.0$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Con gamma = 2.0
#gamma = 2.0
#w1 = np.sqrt(pow(omo,2) - pow(gamma,2))
#q = [np.pi / 3, np.pi / 5, 0, 0]
#p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular
#delta1 = del1(gamma1,w1)
#delta2 = []

#for i in range (2):
    #delta2.append(del2(p[i],p[i+2],wf))

#for i in range (2):
    #A1.append(As(q[i],delta2[0]))
    #A2.append(As(q[i],delta2[1]))

#theta1 = the(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
#theta2 = the(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
#omega1 = ome(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
#omega2 = ome(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])

#Fase
#plt.figure(figsize=(10,10))
#plt.plot(theta1,omega1)
#plt.plot(theta2,omega2)
#plt.show()

#Si A1 = 0

#theta1 = the(t,gamma,b,w1,delta1,A,wf,delta2[0])
#theta2 = the(t,gamma,b,w1,delta1,A,wf,delta2[1])
#omega1 = ome(t,gamma,b,w1,delta1,A,wf,delta2[0])
#omega2 = ome(t,gamma,b,w1,delta1,A,wf,delta2[1])

#Fase
#plt.figure(figsize=(10,10))
#plt.plot(theta1,omega1)
#plt.plot(theta2,omega2)
#plt.show()

#Con gamma = 0.1
gamma = 0.1
w1 = np.sqrt(pow(omo,2) - pow(gamma,2))
q = [np.pi / 3, np.pi / 5, 0, 0]
p = [2 * np.pi / 3, np.pi / 4, 0, 0] #Para solución particular
delta1 = del1(gamma1,w1)
delta2 = []

for i in range (2):
    delta2.append(del2(p[i],p[i+2],wf))

for i in range (2):
    A1.append(As(q[i],delta2[0]))
    A2.append(As(q[i],delta2[1]))

theta1 = the(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,A1[0],w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,A1[1],w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} \\neq 0$, $\\gamma = 0.1$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()

#Si A1 = 0

theta1 = the(t,gamma,b,w1,delta1,A,wf,delta2[0])
theta2 = the(t,gamma,b,w1,delta1,A,wf,delta2[1])
omega1 = ome(t,gamma,b,w1,delta1,A,wf,delta2[0])
omega2 = ome(t,gamma,b,w1,delta1,A,wf,delta2[1])

#Fase
plt.figure(figsize=(10,10))
plt.plot(theta1,omega1,'k',label='$\\dot \\theta_{1}(\\theta_{1})$', linewidth=2)
plt.plot(theta2,omega2,'darkred',label='$\\dot \\theta_{2}(\\theta_{2})$', linewidth=2)
plt.title('Espacio fase (dos condiciones iniciales), $A_{H} = 0$, $\\gamma = 0.1$')
plt.ylabel('Frecuencia angular $(\\dot \\theta)$')
plt.xlabel('Ángulo $(\\theta)$')
plt.legend()
plt.grid(True,color='slategray')
plt.show()
```

**2.- Ecuación no lineal no forzada sin amortiguamiento.**

**a)** Para un ángulo inicial de 70 grados en reposo y una frecuencia natural unitaria, realiza una gráfica donde compares la solucion numérica del ángulo y su velocidad con aquella de la ecuación lineal usando las mismas condiciones iniciales.

**b)** Construye el espacio fase para una frecuencia natural unitaria, usando trayectorias que partan del reposo con un ángulo inicial que vaya de $\dfrac{-\pi}{2}$ a $\dfrac{\pi}{2}$, en intervalos de $\dfrac{\pi}{18}$. Suma dos trayectorias que den vueltas, una en contra de las manecillas del reloj y otra a favor.

*Solución numérica*

Con el análisis mecánico, se puede demostrar que bajo las condiciones impuestas en el enunciado, la expresión no lineal para oscilaciones no forzadas y sin amortiguamiento es la siguiente:

$$\ddot{x} + senx = 0$$ (IX)

La solución que corresponde a la expresión *(IX)* se resolverá con la distribución **ODEINT,** y como la instrucción lo requiere, se resolverán dos ecuaciones

$$\omega = \dot{\theta}$$
$$\dot{\omega} = -sen\theta$$

La estructura de la instrucción es la siguiente:

$$ y = odeint(modelo, y_{0}, t)$$

donde **modelo** es el nombre de la función que devuelve valores derivados en los valores $y$ y $t$ solicitados como $D_{t}y = modelo (y, t),$ $y_{0}$ son las condiciones iniciales de los estados diferenciales, **t** los puntos de tiempo en los que se debe informar la solución. Los puntos internos adicionales a menudo se calculan para mantener la precisión de la solución, pero no se informan. 

La solución numérica se realizará con este método.

**Solución a la ecuación diferencial: lineal y no lineal**

*ED: $\theta(t)$.*

```python
import numpy as np
import math
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from matplotlib import style

# Definición de ED:
def nlin(y, t, b, omo):
    theta, omega = y
    dnlindt = [omega, -b*omega - pow(omo,2)*np.sin(theta)]
    return dnlindt

def lin(y, t, b, c):
    theta, omega = y
    dlindt = [omega, -b*omega - pow(omo,2)*(theta)]
    return dlindt

#Inciso a
#Variables
omo = 1
b = 0
sol = []
soln = []
soll = []
t = np.linspace(0,20,10000)

#Condiciones iniciales 
y0 =[7 * np.pi/18,0]

soln.append(odeint(nlin,y0,t,args = (b,omo)))
soll.append(odeint(lin,y0,t,args = (b,omo)))
    
i = 0

#Figuras
plt.figure(figsize=(25,5))
for i in range(len(soln)):
    plt.plot(t,soln[i][:, 0],'k',label='Solución no lineal  $(\\theta_{NL})$', linewidth=2)
    plt.plot(t,soll[i][:, 0],'darkred',label='Solución lineal $(\\theta_{L})$', linewidth=2)
    plt.title('Solución lineal y no lineal del ángulo en función del tiempo $(\\theta(t))$')
    plt.ylabel('Ángulo $\\theta(t)$')
    plt.xlabel('Tiempo $(t)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()

i = 0
plt.figure(figsize=(25,5))
for i in range(len(soln)):
    plt.plot(t,soln[i][:, 1],'k',label='Solución no lineal de la velocidad $(\\dot \\theta_{NL})$', linewidth=2)
    plt.plot(t,soll[i][:, 1],'darkred',label='Solución lineal $(\\dot \\theta_{L})$', linewidth=2)
    plt.title('Solución lineal y no lineal de la velocidad $(\\dot \\theta(t))$')
    plt.ylabel('Velocidad $\\dot \\theta(t)$')
    plt.xlabel('Tiempo $(t)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()

#Inciso b
#Para las trayectorias
th = np.linspace(-90,90,10)
sol_fr = []
sol_cr = []
t_r = np.linspace(0,500,10000)

for i in range (len(th)):
    yo_r = [th[i]*np.pi/180,0]
    if pow(-1,i) < 0:
        sol_fr.append(odeint(nlin,yo_r,-t_r,args = (b,omo)))
    else:
        sol_cr.append(odeint(nlin,yo_r,t_r,args = (b,omo)))

#Espacio fase, con trayectorias a favor y en contra de las manecilas del reloj
i = 0
plt.figure(figsize=(10,10))
for i in range(len(sol_fr)):
    plt.plot(sol_fr[i][:,0],sol_fr[i][:, 1],label='$\\theta \\circlearrowright$', linewidth=2)
    plt.plot(sol_cr[i][:,0],sol_cr[i][:, 1],label='$\\theta \\circlearrowleft$', linewidth=2)
    plt.title('Espacio fase para diferentes trayectorias')
    plt.ylabel('Velocidad angular $\\dot \\theta$')
    plt.xlabel('Ángulo $(\\theta)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()

#Para la suma de trayectorias    
plt.figure(figsize=(20,20))
for i in range (len(sol_fr)):
    plt.subplot(3,2,i + 1)
    plt.plot(sol_fr[i][:,0] + sol_cr[i][:,0],sol_fr[i][:, 1] + sol_cr[i][:,1],'k',label='$\\theta(0)={0}$'.format(th[i]), linewidth=2)
    plt.title('Espacio fase para la suma de diferentes trayectorias')
    plt.ylabel('Velocidad angular $\\dot \\theta$')
    plt.xlabel('Ángulo $(\\theta)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()
```

Una forma de obtener el **espacio fase** es considerando las expresiones obtenidas para la solución a la ecuación diferencial de segundo orden para la ecuación no lineal, por la regla de la cadena: $\dfrac{d\omega}{d\theta} = \dfrac{\dot{\omega}}{\dot{\theta}} = \dfrac{-\omega_{o}²sen\theta}{\omega}$, resolviendo para $\omega(\theta)$, con $\omega = \dot{\theta}$, obtenemos la siguiente expresión

$$\omega² = 2 \omega_{o}²(cos\theta - 1)$$

Es una de las maneras en las que se puede ilustrar el **espacio fase**. A continuación se ilutra.

**4.- Ecuación no lineal completa.**

**a)** Usando $\mu = 3\pi/2, \omega = 2\pi, \omega_{o} = 3\pi$, grafica la solución para cinco casos de **A** alrededor de **1.05.** ¿Qué pasa con las soluciones? ¿Qué ocurre si **A = 2**?

**b)** Obten las trayectorias en el espacio fase de los casos del inciso anterior.

```python
# Definición de ED:
def pend(y,t,gamma,omo,A,omf):
    theta,omega = y 
    dydt = [omega,(A * np.cos(omf * t) - (2 * gamma * omega) - (pow(omo,2) * np.sin(theta)))]
    return dydt

#Variables
gamma = (3 * np.pi) / 2
omo = 3 * np.pi
omf = 2 * np.pi
A = [0.5,0.8,1.0,1.2,1.5,2]
sol = []
t = np.linspace(0,10,10000)

#Condiciones iniciales 
y0 =[np.pi/3,0]

for i in range (6):
    sol.append(odeint(pend,y0,t,args = (gamma,omo,A[i],omf)))
    
i = 0
#Figuras
plt.figure(figsize=(25,5))
for i in range(6):
    plt.plot(t,sol[i][:, 0],label='$A={0}$'.format(A[i]), linewidth=2)
    plt.title('Trayectoria $\\theta(t)$ para diferentes valores de A')
    plt.ylabel('Ángulo $(\\theta)$')
    plt.xlabel('Tiempo $(t)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()

i = 0
plt.figure(figsize=(25,5))
plt.ylim(-0.05,0.05)
for i in range(6):
    plt.plot(t,sol[i][:, 0],label='$A={0}$'.format(A[i]), linewidth=2)
    plt.title('Trayectoria $\\theta(t)$ para diferentes valores de A')
    plt.ylabel('Ángulo $(\\theta)$')
    plt.xlabel('Tiempo $(t)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()

i = 0
plt.figure(figsize=(25,25))
for i in range (len(sol)):
    plt.subplot(3,3,i + 1)
    plt.plot(sol[i][:, 0],sol[i][:,1],'k',label='$A={0}$'.format(A[i]), linewidth=2)
    plt.title('Trayectoria $\\theta(t)$ para diferentes valores de A')
    plt.ylabel('Ángulo $(\\theta)$')
    plt.xlabel('Tiempo $(t)$')
    plt.legend()
    plt.grid(True,color='slategray')
plt.show()
```

Con $A = 2$, el sistema corresponde en menor tiempo al oscilador armónico, es decir, la magnitud de la fuerza es mayor y las frecuencias entrar con mayor rapidez al estado transitorio. Entonces, entre mayor sea la amplitud, la solución corresponerá más rápidamente al oscilador armónico.
